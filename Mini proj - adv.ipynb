# Final Mini-Project: Gaming vs Academic Performance
**Dataset:** Gaming vs Academic Performance  
**Source:** https://www.kaggle.com/datasets/aiexplorer77/gaming-vs-academic-performance  
**Goal:** Apply core pandas skills — importing, inspecting, selecting, filtering, cleaning, grouping, summarising and exporting data.

#### Operation 1 — Import Libraries and Load the Dataset

```python
import pandas as pd
import numpy as np

# Read CSV into a DataFrame
df = pd.read_csv('Gaming_Academic_Performance.csv')
print('Dataset loaded successfully.')
print(f'Shape: {df.shape[0]} rows, {df.shape[1]} columns')
```

#### Operation 2 — Inspect the Data

```python
# View the first 5 rows
df.head()
```

```python
# Check column names and data types
print('Columns:', df.columns.tolist())
print()
print(df.dtypes)
```

#### Operation 3 — Summary Statistics

```python
# Summary statistics for all numeric columns
df.describe()
```

#### Operation 4 — Check and Handle Missing Values

```python
# Check for missing values
print('Missing values per column:')
print(df.isnull().sum())
print(f'\nTotal missing: {df.isnull().sum().sum()}')
```

```python
# Simulate 5 missing values in 'grades' to demonstrate fillna()
df_demo = df.copy()
df_demo.loc[[10, 50, 200, 500, 1000], 'grades'] = np.nan
print('Simulated missing values in grades:', df_demo['grades'].isnull().sum())

# Fill with median
median_grade = df_demo['grades'].median()
df_demo['grades'] = df_demo['grades'].fillna(median_grade)
print(f'After filling with median ({median_grade:.2f}):', df_demo['grades'].isnull().sum())
```

```python
# Remove duplicate rows
df_clean = df.drop_duplicates()
print(f'Rows after removing duplicates: {len(df_clean)} (removed {len(df) - len(df_clean)})')
```

#### Operation 5 — Select Specific Columns

```python
# Select only key columns
key_cols = df[['student_id', 'gaming_hours', 'study_hours', 'grades', 'stress_level']]
key_cols.head()
```

#### Operation 6 — Filter Rows (Boolean Indexing)

```python
# Students with high addiction AND low grades
at_risk = df[(df['addiction_score'] >= 15) & (df['grades'] < 60)]
print(f'At-risk students (high addiction + low grades): {len(at_risk)}')
at_risk[['student_id', 'gaming_hours', 'addiction_score', 'grades', 'stress_level']].head(10)
```

```python
# Students gaming more than 6 hours per day
heavy_gamers = df[df['gaming_hours'] > 6]
print(f'Heavy gamers (>6 hrs/day): {len(heavy_gamers)} students ({len(heavy_gamers)/len(df)*100:.1f}%)')
```

#### Operation 7 — groupby() and Aggregation

```python
# Average grades, gaming hours and study hours by stress level
stress_summary = df.groupby('stress_level')[['grades', 'gaming_hours', 'study_hours']].mean().round(2)
stress_summary.columns = ['Avg Grade', 'Avg Gaming Hrs', 'Avg Study Hrs']
print('Average outcomes by stress level:')
stress_summary
```

```python
# Average grades by gaming genre
genre_grades = df.groupby('gaming_genre')['grades'].agg(['mean', 'median', 'count']).round(2)
genre_grades.columns = ['Mean Grade', 'Median Grade', 'Count']
genre_grades.sort_values('Mean Grade', ascending=False)
```

#### Operation 8 — Create Derived Columns Using apply()

```python
# Gaming-to-study ratio
df_clean['gaming_study_ratio'] = (df_clean['gaming_hours'] / df_clean['study_hours']).round(3)

# Grade band using apply()
def grade_band(g):
    if g >= 85:
        return 'High (85+)'
    elif g >= 65:
        return 'Medium (65-84)'
    else:
        return 'Low (<65)'

df_clean['grade_band'] = df_clean['grades'].apply(grade_band)
print('Grade band distribution:')
print(df_clean['grade_band'].value_counts())
```

#### Operation 9 — value_counts() and Correlation

```python
# Frequency counts for categorical columns
print('Stress level distribution:')
print(df['stress_level'].value_counts())
print()
print('Gaming genre distribution:')
print(df['gaming_genre'].value_counts())
```

```python
# Correlation of numeric columns with grades
numeric_cols = ['gaming_hours', 'study_hours', 'sleep_hours', 'attendance',
                'addiction_score', 'reaction_time_ms', 'social_activity', 'device_usage']

correlations = df[numeric_cols + ['grades']].corr()['grades'].drop('grades').sort_values(ascending=False).round(3)
print('Correlation with grades (strongest to weakest):')
print(correlations)
```

#### Operation 10 — Export Cleaned Dataset

```python
# Export cleaned dataset with new columns to CSV
df_clean.to_csv('Gaming_Academic_Performance_cleaned.csv', index=False)
print('Exported to: Gaming_Academic_Performance_cleaned.csv')
print(f'Rows: {len(df_clean)}, Columns: {len(df_clean.columns)}')
print('New columns added:', [c for c in df_clean.columns if c not in df.columns])
```

---
## Three Key Findings

**Finding 1 — Study hours are the strongest positive predictor of grades.**  
The correlation analysis shows `study_hours` has the highest positive correlation with grades.

**Finding 2 — High stress is associated with lower grades and more gaming.**  
High-stress students have lower average grades and higher average gaming hours compared to low-stress students.

**Finding 3 — Addiction score and gaming hours negatively correlate with grades.**  
Both variables show negative correlations with grades. The at-risk filter identifies students where excessive gaming links to poor academic outcomes.
